```python
#*******DEPENDENCIES AND SETUP**********

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import csv
import json

#import API key
from api_keys import api_key

#Incorporated citypy to determine city base on lat/long
from citipy import citipy

#output file (csv)
output_data_file = "output_data/cities.csv"

#Range of latitudes and longitudes
# lat_range=(-90,90)
# lng_range=(-180,180)

#settings = {"units": "imperial", "appid": api_key}

```

```python
#*******GENERATE CITIES LIST***************

#list for holding lat_long and cities
lat_lngs=[]
lat_lng_list=[]
cities=[]

#create a set of random lat and long combinations
lats=np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs=np.random.uniform(low=-180.000, high=180.000,size=1500)
lat_lngs=zip(lats, lngs)

#identify nearest city for each lat, lng combo
for lat_lng in lat_lngs:
    city=citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    #if the city is unique, add it to cities list
    if city not in cities:
        cities.append(city)
        lat_lng_list.append(lat_lng)

        
#print the city count to confirm sufficient count
len(cities)
#cities
```

```python
#**********PERFORM API CALLS*********

# Get data for each city in cities_df
#city_weather_dict = {}

url= "https://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
 
# Build partial query URL
query_url = url + "appid=" + api_key + "&units=" + units + "&q=" 
```

```python
cities_found = []
cloudiness = []
country_codes=[]
dates = []
humidity = []
latitude = []
longitude = []
temperature = []
wind_speed = []


print("Begin Data Processing")
print("--------------------------------------------")

for x in range(len(cities)):
    print(f"Processing Record {x} for city: {cities[x]}")
    print(f"{url}&q={cities[x]}")
    
    #get the cities
    cities_response=requests.get(query_url+str(cities[x])).json()
    
    try:
        cities_found.append(cities_response['name'])
        humidity.append(cities_response['main']['humidity'])
        cloudiness.append(cities_response['clouds']['all'])
        country_codes.append(cities_response['sys']['country'])
        dates.append(cities_response['dt'])
        temperature.append(cities_response['main']['temp_max'])
        wind_speed.append(cities_response['wind']['speed'])
        latitude.append(cities_response['coord']['lat'])
        longitude.append(cities_response['coord']['lon'])
    except:
        print(f'Sorry, unable to locate {cities[x]}')
    
    
    
    
print("--------------------------------------------")
print("Data Processing Complete")
print("--------------------------------------------")


#output data to csv

```

```python
#**********PUT CITIES IN A DF*********
city_df=pd.DataFrame({
    'City':cities_found
})
#city_df.head()
```
